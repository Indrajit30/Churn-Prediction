```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from split import split_data
from sklearn.linear_model import LogisticRegression
from error_analysis import classification_summary
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv("/content/drive/MyDrive/Customer_Churn_Prediction/Telco-Customer-Churn-Final.csv")
```

```python
x = df.drop(columns="Churn", errors="ignore")
y = df["Churn"].astype(int)
```

```python
x_train, x_val, x_test, y_train, y_val, y_test = split_data(x,y)
```

```python
x_val
```

```python
logistic = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler()),
    ("model", LogisticRegression(
        max_iter=5000,
        class_weight="balanced",
        solver="liblinear"
    ))
])
```

```python
_ = logistic.fit(x_train, y_train)
```

```python
proba = logistic.predict_proba(x_val)[:, 1]
threshold = 0.5
pred = (proba >= threshold).astype(int)
```

```python
print(y_val.shape)
print(pred.shape)
```

```python
print(classification_summary(y_val, pred, proba))
```
